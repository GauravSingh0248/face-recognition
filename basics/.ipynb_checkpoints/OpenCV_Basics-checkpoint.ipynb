```python
# create a camera object 
cam=cv2.VideoCapture(0)
# read image from camera object 
while True: #10 raise to the power 8 iamges per second 
    success,img=cam.read()
    if not success:
        print("reading from camera failed")
    cv2.imshow("Image window ",img)
    # cv2.waitKey(0)  stay for indefinite time 
    # cv2.waitKey(4000) #stays for 4000 milli seconds
    cv2.waitKey(1) #pause here for 1 ms before you read the next image

```
